```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
```

```python
X, y = load_digits(return_X_y=True)
```

```python
norm = StandardScaler()
```

```python
x_std = norm.fit_transform(X)
```

```python
x_std.shape
```

```python
pca = PCA(n_components=2)

X_transformada = pca.fit_transform(x_std)

X_transformada = pd.DataFrame(X_transformada, columns = ['C1', 'C2'])
X_transformada['label'] = y
```

```python
X_transformada.head()
```

```python
plt.figure(figsize=(10,10))
colores = ['b.', 'g.', 'r.', 'c.', 'm.', 'y.', 'k.', 'w.','y>', 'r>']
for i in range(0,10):
    plt.plot(X_transformada['C1'][X_transformada['label']==i],
             X_transformada['C2'][X_transformada['label']==i], colores[i], label=i)  
plt.legend()
plt.show()
```

```python
pca3 = PCA(n_components=3)
X_transformada = pca3.fit_transform(x_std)
X_transformada = pd.DataFrame(X_transformada, columns = ['C1', 'C2', 'C3'])
X_transformada['label'] = y
```

```python
fig = px.scatter_3d(X_transformada, x='C1', y='C2', z='C3',
              color='label')
fig.show()
```
